# What time of the year, on average, is the best time to host summer parties in Philadelphia? 

For the best summer parties, whether it's a pool party or an outdoor barbecue, two important factors are necessary: it's warm outside, and there is no rain. Sometimes, people need to plan their summer parties in advance to accommodate any travel plans. l

To give Philadelphia's residents a better idea of when to host these summer parties, I analyzed weather data from 2010 to 2025 to determine what days tend to have the warmest temperatures or the most rainfall.

### What tends to be the warmest time of the year in Philadelphia?

![](sunny.jpg)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weather_df = pd.read_csv('philadelphia_weather_2005_to_2025.csv')
weather_df = weather_df.assign(date=pd.to_datetime(weather_df['date']))
```

```python
weather_df = weather_df[weather_df["date"].dt.year >= 2010]
```

```python
warmest_days = weather_df.loc[
    weather_df.groupby(weather_df["date"].dt.year)["high"].idxmax()
]

warmest_days[["date", "high"]]
```

From this table, you can see that Philadelphia mostly sees high temperatures between 96 and 103. Most of these dates occur in July and August. When you take the average of these days, you get July 16. 

This data shows that Philadelphia residents can trust July and August for hosting their summer parties and can expect high and warm temperatures. Specifically, July 16 seems like the safest bet for a pool party, since history shows that the warmest temperatures occur around this date every year.

### What tends to be the rainiest time of the year in Philadelphia?

![](rain.jpg)

```python
rainiest_days = weather_df.loc[
    weather_df.groupby(weather_df["date"].dt.year)["rain"].idxmax()
]

rainiest_days[["date", "rain"]]
```

From this table, we can see that September and the fall months tend to have the most rainfall, with up to 8 inches of rain in August. The rainfall data is more scattered, though, and is not concentrated to one specific month. 

However, when we take the average of these dates, we get July 30. 

From this, we can determine that hosting summer parties is not a good idea toward the end of July. Additionally, hosting outdoor parties is also not a good idea at the beginning of fall. Not just because of the weather, but also because it could rain.

### Conclusion

After reviewing this data, we can determine that the best time of year to host a summer party is mid-July. This time of year has the warmest weather, according to the average data. Mid-July also swiftly avoids any rainfall. This is perfect for any Fourth of July barbecue!
